# Computing a dataframe with empirical probalities of WNV | # in trap & species


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
import time
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

from sklearn.cluster import KMeans, k_means
from sklearn.metrics import silhouette_score

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 999)


%matplotlib inline
```

```python
kaggle_train = pd.read_csv('../data/train.csv')
kaggle_test = pd.read_csv('../data/test.csv')
```

```python
X_k_train = kaggle_train[['Date', 'Species', 'Trap', 'Latitude', 'Longitude', 'NumMosquitos','WnvPresent']].copy()
X_cols = list(X_k_train.columns.drop(['WnvPresent','NumMosquitos']))
X_cols
```

```python
X_k_train['Date'] = pd.to_datetime(kaggle_train['Date'])
```

```python
X_k_train.head()
```

```python
spec = list(set(X_k_train['Species']))

df = pd.DataFrame(0, index = range(51), columns = spec)

df.head()
```

```python
def add_to_df(df_in, species, yn_var):
    df_in.loc[yn_var,species] += 1
    return(df_in)

for each in range(len(X_k_train)):  
    df = add_to_df(df, X_k_train.loc[each,'Species'], X_k_train.loc[each,'NumMosquitos'])


df.head()       # number of occurance of species / trap in the training data
```

```python
def add_to_df2(df_in, species, NumM, Wnv):
    df_in.loc[NumM,species] += Wnv
    return(df_in)

df2 = pd.DataFrame(0, index = range(51), columns = spec)
for each in range(len(X_k_train)):  
    df2 = add_to_df2(df2, X_k_train.loc[each,'Species'], X_k_train.loc[each,'NumMosquitos'],X_k_train.loc[each,'WnvPresent'])


df2.head()           # number of WNV Present occurances by Species and Number in trap
```

```python
df3 = df2.copy()
for i in range(len(df2)):
    for j in df2.columns:
        if df.loc[i,j] == 0:
            df3.loc[i,j] = 0
        else:
            df3.loc[i,j] = df2.loc[i,j] / df.loc[i,j]
            
df3.head()     # Empirical probabilities of WNV present given Species and Number of Mosquitos in trap
```

```python
df3.describe()
```

```python
print('probabilities by species')
df2.sum()/df.sum()                     # probability of WNV given species
```

```python
# The empirical probabilities are not smooth (ie, data is not robust)
# These are the probabilities grouped by 5 durations of Number of Mosquitos in the trap
# Initially I would assume that this would increase with # of mosquitos in the trap.

print('probabilies by groupings of 5 mosquitos per trap')
for col in ['CULEX PIPIENS', 'CULEX RESTUANS', 'CULEX PIPIENS/RESTUANS']:
    print(col,":")
    for j in range(1, 51, 5):
        if df.loc[j:j+5,col].sum() == 0:
            x = 0
        else:
            x = df2.loc[j:j+5,col].sum()/df.loc[j:j+5,col].sum()
        print(j, '-', j+5, x)
```

```python
# Overal probability that there is WNV given at least one mosquito in a trap.

print('probabilies for all observations in total')
df2.sum().sum()/df.sum().sum()
```

### Making probability dataframe (probs).  Since the probabilities are not smooth, I just visually inspected the probabilities above (5 groups of # of mosquitos) and then made a smooth line of the probabilities.  

NB:  This could probably be fit using a linear regression to improve the accuracy.

```python
probs = pd.DataFrame(0, index = range(51), columns = set(kaggle_test['Species']))
```

```python
vec = [0]
vec.extend(np.linspace(.01, .2, 40))
vec.extend(np.linspace(.2,.12,10))
probs['CULEX PIPIENS/RESTUANS'] = vec
```

```python
vec = [0]
vec.extend(np.linspace(.02, .32, 40))
vec.extend(np.linspace(.32,.12,10))
probs['CULEX PIPIENS'] = vec
```

```python
vec = [0]
vec.extend(np.linspace(.01, .03, 30))
vec.extend(np.linspace(.03,.05,5))
vec.extend(np.linspace(.05,.07,10))
vec.extend(np.linspace(.05,.025,5))
probs['CULEX RESTUANS'] = vec
```

```python
vec = [0]
vec.extend(np.linspace(.05, .05, 50))
probs['UNSPECIFIED CULEX'] = vec
```

### This section models the data to predict the number of mosquitos in the Test data.

```python
X_k_test = kaggle_test[['Id', 'Date', 'Species', 'Trap', 'Latitude', 'Longitude']].copy()
X_k_test['Date'] = pd.to_datetime(X_k_test['Date'])
#X_k_test.set_index('Date', drop = True, inplace = True)
```

```python
X = X_k_train[['Date', 'Species', 'Latitude', 'Longitude', 'NumMosquitos']].copy()
X['week'] = X['Date'].dt.strftime('%W')
y = X[['NumMosquitos']]
```

```python
X.drop(['Date','NumMosquitos'], axis = 1, inplace=True)
```

```python
X = pd.get_dummies(X, columns = ['Species'])
```

```python
X['Species_UNSPECIFIED CULEX'] = 0         # Add this in to make train and kaggle_test equal in columns
```

```python
Test = X_k_test[['Date','Species', 'Latitude', 'Longitude']].copy()
Test['week'] = Test['Date'].dt.strftime('%W')
Test.drop(['Date'], axis = 1, inplace = True)
```

```python
Test=pd.get_dummies(Test, columns = ['Species'])
```

```python
Test.columns      # basically, I am modeling lat / long, week, and species
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1929)
```

```python
# KNN didn't really work out great.

from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

for i in range(2,40):
    knn = KNeighborsRegressor(n_neighbors = i)
    model = knn.fit(X_train, y_train)
    yhat = model.predict(X_test)
    print('i = ', i)
    print(r2_score(y_train, knn.predict(X_train)))
    print(r2_score(y_test, yhat))
    print()
```

```python
# okay fit but overall not very accurate with LR

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

print(r2_score(y_train,lr.predict(X_train)))
print(r2_score(y_test,lr.predict(X_test)))
```

```python
#Gradient Boost have the best score so far.

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score


gb = GradientBoostingRegressor()
model = gb.fit(X_train, y_train)
yhat = model.predict(X_test)
print(r2_score(y_train, knn.predict(X_train)))
print(r2_score(y_test, yhat))
print()
```

```python
yhat = gb.predict(Test)
```

```python
# manually force the predictions to conform to integers in the range of [0,50]
int_yhat = []
for each in yhat:
    if each <0.5:
        x = 0
    elif each > 51.5:
        x = 50
    else:
        x = int(each)
    int_yhat.append(x)
```

```python
len(kaggle_test)
```

```python
len(int_yhat)
```

```python
# Add Mosquito project for each trap
# look up probabilities

kaggle_test['NumMosquitos'] = int_yhat
kaggle_test['WnvPresent'] = 0

for each in range(len(kaggle_test)):
    sp = kaggle_test.loc[each, 'Species']
    num = kaggle_test.loc[each, 'NumMosquitos']
    kaggle_test.loc[each, 'WnvPresent'] = probs.loc[num,sp]
```

```python
# This ties the best submission exactly
# Put in a probability for the Unspecified
# Totally useless exercise

kaggle_test['NumMosquitos'] = int_yhat
kaggle_test['WnvPresent'] = 0

for each in range(len(kaggle_test)):
    sp = kaggle_test.loc[each, 'Species']
    num = kaggle_test.loc[each, 'NumMosquitos']
    kaggle_test.loc[each, 'WnvPresent'] = probs.loc[num,sp]
    if sp == 'UNSPECIFIED CULEX':
        kaggle_test.loc[each, 'WnvPresent'] = .01
        
```
